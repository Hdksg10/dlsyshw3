```python
import numpy as np
m = 128
n = 128
p = 128

test_a = np.zeros((m, n))
test_b = np.ones((n, p))

for i in range(m):
    for j in range(n):
        test_a[i][j] = i*n+j

# init test_b array from 0 to 255
for i in range(n):
    for j in range(p):
        test_b[i][j] = i*p+j
```

```python
print(test_a)
```

```python
print(test_b)
```

```python
print(test_b[:, 0])
```

```python
print(test_a @ test_b)
```
